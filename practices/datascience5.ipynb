```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

array_x = np.arange(-3,4)
array_f = -array_x*array_x
print('array_x=',array_x)
print('array_f=',array_f)

plt.plot(array_x,array_f)
```

```python
array_x = np.arange(-3,3.1,0.2)
array_f = -array_x*array_x
plt.plot(array_x,array_f)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

mu = 4
array_x = np.arange(-3+mu,3.1+mu,0.2)
array_f = -(array_x-mu)*(array_x-mu)
plt.plot(array_x,array_f)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
mu = 4
array_x = np.arange(-3,3.1+mu,0.2)
array_f1 = -array_x*array_x
array_f2 =-(array_x-mu)*(array_x-mu)
plt.ylim([-9,1])
plt.plot(array_x,array_f1)
plt.plot(array_x,array_f2)
```

```python
mu = 4
array_x = np.arange(-3,3.1,0.2)
array_x1 = np.arange(-3+mu,3.1+mu,0.2)
array_f1 = -array_x*array_x
array_f2 =-(array_x1-mu)*(array_x1-mu)
plt.ylim([-9,1])
plt.plot(array_x,array_f1)
plt.plot(array_x1,array_f2)
```

```python
sigma = 2
array_x = np.arange(-3*sigma,3.01*sigma,0.2)
array_f = -(array_x/sigma)*(array_x/sigma)
plt.plot(array_x,array_f)
```
